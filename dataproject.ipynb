# Read and clean data

Read population data in population.csv and clean it:

```python
# Load the data
total_pop = pd.read_csv('population.csv')

# Drop unwanted columns
pop = total_pop.drop(['Series Name', 'Series Code', '2017 [YR2017]','2018 [YR2018]', '2019 [YR2019]', '2020 [YR2020]'], axis = 1)

# Filter data for the desired countries
countries = ["Afghanistan", "Bhutan", "Bangladesh", "India", "Myanmar", "Maldives", "Nepal", "Pakistan", "Sri Lanka"]

# Create a dataframe
Southern_Asia = pop[pop["Country Name"].isin(countries)]

# Rename the column
Southern_Asia = Southern_Asia.rename(columns={'2021 [YR2021]': 'Population'}).reset_index()

# Create a new column
Southern_Asia['Year'] = 2021

# Drop the series number
Population = Southern_Asia.drop('index', axis=1)

# Show the data
Population.head(9)
```

Read total GDP data in gdp.csv and clean it:

```python
# Load the data
total_gdp = pd.read_csv('gdp.csv')

# Drop unwanted columns
gdp = total_gdp.drop(['Series Name', 'Series Code', '2017 [YR2017]','2018 [YR2018]', '2019 [YR2019]', '2020 [YR2020]'], axis = 1)

# Define a list of countries to filter
countries = ["Afghanistan", "Bhutan", "Bangladesh", "India", "Myanmar", "Maldives", "Nepal", "Pakistan", "Sri Lanka"]

# Filter dataframe to include only specified countries
Southern_Asia = gdp[gdp["Country Name"].isin(countries)]

#Rename the column
Southern_Asia = Southern_Asia.rename(columns={'2021 [YR2021]': 'Total GDP'}).reset_index()

# Create a new column
Southern_Asia['Year'] = 2021

# Drop series number
GDP_Total = Southern_Asia.drop('index', axis=1)

# Show the data
GDP_Total.head(9)
```

Read GDP per Capita data in gdp per capita.csv and clean it:

```python
# Load the data
gdp_per_capita = pd.read_csv('gdp per capita.csv')

# Drop unwanted columns
GDP_per_capita = gdp_per_capita.drop(['Series Name', 'Series Code', '2017 [YR2017]','2018 [YR2018]', '2019 [YR2019]', '2020 [YR2020]'], axis = 1)

# Define a list of countries to filter
countries = ["Afghanistan", "Bhutan", "Bangladesh", "India", "Myanmar", "Maldives", "Nepal", "Pakistan", "Sri Lanka"]

# Filter dataframe to include only specified countries
Southern_Asia = GDP_per_capita[GDP_per_capita["Country Name"].isin(countries)]

# Rename the column
Southern_Asia = Southern_Asia.rename(columns={'2021 [YR2021]': 'GDP Per Capita'}).reset_index()

# Create a new column and initialize it with 2021
Southern_Asia['Year'] = 2021

# Drop the series number
GDP_per_cap = Southern_Asia.drop('index', axis=1)

# Show the data
GDP_per_cap.head(9)
```

# Explore dataset

Here we put three plots about population, total GDP and GDP per Capita respectively 

Plot population data in line graph:

```python
import re

# Get the population values and years as arrays
population = Population['Population'].apply(lambda x: re.sub(r'[^\d.]+', '', x))
population = pd.to_numeric(population, errors='coerce')
countries = Population['Country Name'].values

# Create a new figure and set the title
plt.figure(figsize=(10, 6))
plt.title('Population in Southern Asia Countries')

# Plot the bar chart with population on the y-axis and countries on the x-axis
plt.plot(countries, population)

# Set the x-axis and y-axis labels
plt.xlabel('Country')
plt.ylabel('Population')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Show the chart
plt.show()
```

Plot total GDP data in scatter graph:

```python
import re
# Get the population values and years as arrays
GDP_total = GDP_Total['Total GDP'].apply(lambda x: re.sub(r'[^\d.]+', '', x))
GDP_total = pd.to_numeric(GDP_total, errors='coerce')
countries = GDP_Total['Country Name'].values

# Create a new figure and set the title
plt.figure(figsize=(10, 6))
plt.title('Total GDP in Southern Asia Countries')

# Plot the bar chart with population on the y-axis and countries on the x-axis
plt.scatter(countries, GDP_total)

# Set the x-axis and y-axis labels
plt.xlabel('Country')
plt.ylabel('Total GDP')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Show the chart
plt.show()
```

Plot GDP per Capita data in bar chart:

```python
import re

# Get the population values and years as arrays
Gdp_per_cap = GDP_per_cap['GDP Per Capita'].apply(lambda x: re.sub(r'[^\d.]+', '', x))
Gdp_per_cap = pd.to_numeric(Gdp_per_cap, errors='coerce')
countries = GDP_per_cap['Country Name'].values

# Create a new figure and set the title
plt.figure(figsize=(10, 6))
plt.title('GDP Per Capita in Southern Asia Countries')

# Plot the bar chart with population on the y-axis and countries on the x-axis
plt.bar(countries, Gdp_per_cap)

# Set the x-axis and y-axis labels
plt.xlabel('Country')
plt.ylabel('GDP Per Capita')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Show the chart
plt.show()
```

# Merge datasets

Firstly, we merge three datasets in one frame:

```python
# Merge the datasets on 'Country Name' and 'Year'
merged_data_1 = pd.merge(Population, GDP_Total, on=['Country Name', 'Year'])
merged_data_2 = pd.merge(merged_data_1, GDP_per_cap, on=['Country Name', 'Year'])

# Drop the repeated column
final = merged_data_1.drop(columns=['Country Code_y'])
final = merged_data_2.drop(columns=['Country Code_y'])

# Drop one of the 'Country Code' columns
final = final.drop(columns=['Country Code'])

# Rename the columns
final = final.rename(columns={'Country Code_x': 'Country Code', 'GDP per capita': 'GDP per Capita'})

# Reorder the columns
final = final[['Country Name', 'Country Code', 'Population', 'Total GDP', 'GDP Per Capita', 'Year']]

# Show the merged data
final.head(9)
```

Next, we plot the data in line graph to analysis:

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats

# Convert the population and GDP values to numeric
final['Population'] = pd.to_numeric(final['Population'], errors='coerce')
final['Total GDP'] = pd.to_numeric(final['Total GDP'], errors='coerce')
final['GDP Per Capita'] = pd.to_numeric(final['GDP Per Capita'], errors='coerce')

# Set the figure size and style
plt.figure(figsize=(10, 6))
plt.style.use('seaborn-darkgrid')

# Create a single subplot with two y-axes
ax = plt.subplot()

# Plot the population data for all countries on the left y-axis
ax.plot(final['Country Name'], final['Population'], color='C0', alpha=0.5)

# Add axis labels and a legend for the left y-axis
ax.set_xlabel('Country Name')
ax.set_ylabel('Population')
ax.legend(['Population'], loc='upper right', bbox_to_anchor=(0.98, 0.95))

# Create a second y-axis for GDP
ax2 = ax.twinx()

# Plot GDP for all countries on the right y-axis
ax2.plot(final['Country Name'], final['Total GDP'], color='C1', marker='o')

# Add axis labels and a legend for the right y-axis
ax2.set_ylabel('Total GDP')
ax2.legend(['Total GDP'], loc='upper right', bbox_to_anchor=(0.98, 0.85))

# Add GDP Per Capita data to the plot
ax3 = ax.twinx()
ax3.spines['right'].set_position(('axes', 1.2))  # Move the third y-axis to the right
ax3.plot(final['Country Name'], final['GDP Per Capita'], color='C2', linestyle='--')
ax3.get_yaxis().set_visible(False)

# Add labels for individual data points
for i, val in enumerate(final['GDP Per Capita']):
    ax3.text(i, val, "{:,.2f}".format(val), ha='center')

# Add axis label for the third y-axis
ax3.legend(['GDP Per Capita'], loc='upper right', bbox_to_anchor=(0.98, 0.75))

# Add a title to the entire figure
plt.suptitle('Population vs. Total GDP vs. GDP Per Capita in Southern Asia Countries (2021)')

# Display the graph
plt.show()
```

# Analysis

It seems sense that there is a correlation between population and total GDP because nations with higher populations typically have more developed economies. India is the most populous of these nations in the south and has the largest GDP overall, although its GDP per person is very low. Additionally, the graph demonstrates that some nations with smaller populations have greater total GDPs than nations with bigger populations, demonstrating that other factors besides population size can affect total GDP.

GDP per capita, a measurement of average wealth per person in each nation, is derived by dividing total GDP by population. It's vital to keep in mind that countries with higher GDP may not necessarily have higher GDP per capita. For example, among the selected Southern Asian nations in 2021, the Maldives has the highest GDP per capita, followed by Bhutan and Sri Lanka. The Maldives is a tiny island country in the Indian Ocean with a modest population and a low overall Economy. Afghanistan, on the other side, has the lowest GDP per capita, which may be related to the continuous unrest and instability there.

# Conclusion

To concise, although other factors than population size might also have an impact on total GDP, the link between population and total GDP in Southern Asian countries appears to be generally favorable. GDP per capita is a helpful metric for assessing the average level of wealth per person in each country and can differ significantly even amongst nations with comparable total GDP.
